```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0)
b = torch.tensor(0.0)
def binary_cross_entropy_loss(prediction, target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction,epsilon,1-epsilon)
  return -(target*torch.log(prediction)+ (1-target)*torch.log(1-prediction))
```

```python
z = w*x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))
dy_pred_dz = y_pred*(1-y_pred)
dz_dw = x
dz_db = 1
```

```python
dl_dw = dloss_dy_pred*dy_pred_dz*dz_dw
dl_db = dloss_dy_pred*dy_pred_dz*dz_db
print(f"The value of dl_dw is {dl_dw}")
print(f"The value of dl_db is {dl_db}")
```

```python
x = torch.tensor(6.7, requires_grad=True)
y = torch.tensor(0.0, requires_grad=True)
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
z = w*x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy_loss(y_pred, y)
loss.backward()
print(f"The value of dl_dw is {w.grad}")
print(f"The value of dl_db is {b.grad}")
```

```python
x = torch.tensor([1.0,2.0,3.0], requires_grad=True)
y = (x**2).mean()
y.backward()
x.grad
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2
z = x**3
z.backward()
x.grad
```

```python
x.grad.zero_() # To clear Gradient
```

```python
x = torch.tensor(2.0, requries_grad=True)
x.requires_grad_(False) # To disable Gradient
y = x**2
z = x**3
z.backward()
x.grad
```

```python
x = torch.tensor(2.0, requires_grad=True)
with torch.no_grad(): # Disable gradient
  y = x**2
  y.backward()
```

```python
x = torch.tensor(3.0, requires_grad=True)
z = x.detach() # Now y.backward() will work but y1.backward() will not work
y = x**2
y1 = z**2
```
