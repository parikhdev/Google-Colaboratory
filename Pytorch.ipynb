```python
import time
a = time.time()
size = 10000
m = torch.randint(size = (size, size), low = 0, high = 100, dtype = torch.float32)
n = torch.randint(size = (size, size), low = 0, high = 100, dtype = torch.float32)
torch.matmul(m,n)
b = time.time()
cpu_time = a - b
print(cpu_time)
```

```python
c = time.time()
m = torch.randint(size = (size, size), low = 0, high = 100, dtype = torch.float32, device = device)
n = torch.randint(size = (size, size), low = 0, high = 100, dtype = torch.float32, device = device)
torch.matmul(m,n)
d = time.time()
gpu_time = c - d
print(gpu_time)
```

```python
ratio = cpu_time/gpu_time
print(f"GPU is {ratio} times faster than CPU")
```
